```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as mb
import seaborn as sns
import scipy.stats 
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix,classification_report
```

```python
LData_url='D:/study/s7 files/UG Project Files/Project Code/Final_Year_Project/Data/final data/DATA_3.csv'
LData=pd.read_csv(LData_url)
LData.drop(['Unnamed: 0'],inplace=True,axis=1)
LData.head()
```

```python
Data_modified=LData
del Data_modified['Label']
del Data_modified['FAULT TYPE']
Data_modified
```

```python
scaler=StandardScaler()
scaler.fit(Data_modified.drop('Label_no',axis=1))
scaled_features=scaler.transform(Data_modified.drop('Label_no',axis=1))
scaled_features
```

```python
scaled_df=pd.DataFrame(scaled_features,columns=Data_modified.columns[:-1])
scaled_df.head()
```

```python
X=scaled_df
Y=Data_modified['Label_no']
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=750,stratify = Data_modified['Label_no'], random_state=132)
```

```python
d_tree=DecisionTreeClassifier()
```

```python
d_tree.fit(X_train,Y_train)
```

```python
predictions=d_tree.predict(X_test)
training_predictions=d_tree.predict(X_train)
```

```python
fault_type = ['IR_7_1','IR_14_1','IR_21_1','BL_7_1','BL_14_1','BL_21_1','OR_7_1','OR_14_1','OR_21_1','NORMAL']
print(classification_report(Y_test,predictions,target_names=fault_type))
```

```python
test_confusion_matrix=confusion_matrix(Y_test,predictions)
training_confusion_matrix=confusion_matrix(Y_train,training_predictions)
fault_type = ['IR_7_1','IR_14_1','IR_21_1','BL_7_1','BL_14_1','BL_21_1','OR_7_1','OR_14_1','OR_21_1','NORMAL']
```

```python
mb.figure(1,figsize=(18,12))
sns.heatmap(training_confusion_matrix,annot=True,cmap='cividis',xticklabels=fault_type,yticklabels=fault_type,fmt='d',annot_kws={'size':15})
mb.xlabel('Preditions',fontsize = 16)
mb.ylabel('True',fontsize = 16)
mb.title('Training Confusion Matrix',fontsize = 16)
```

```python
mb.figure(2,figsize=(20,12))
sns.heatmap(test_confusion_matrix,annot=True,cmap='cividis',xticklabels=fault_type,yticklabels=fault_type,annot_kws={'size':15})
mb.xlabel('Preditions',fontsize = 16)
mb.ylabel('True',fontsize = 16)
mb.title('Test Confusion Matrix',fontsize = 16)
```

```python
mb.figure(3,figsize=(18,12))
sns.heatmap(test_confusion_matrix/75,annot=True,cmap='cividis',xticklabels=fault_type,yticklabels=fault_type,fmt='.2f',annot_kws={'size':15})
mb.xlabel('Preditions',fontsize = 16)
mb.ylabel('True',fontsize = 16)
mb.title('Test Confusion Matrix (%)',fontsize = 16)


```
